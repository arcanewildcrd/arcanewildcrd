### Here we are trying to find a relationship between BMI and AgeCategory in relation to kidney disease
### we want to find out if bmi an d age can be responsible for a person getting kidney disease

```python
#sample data with KidneyDisease as string
data = {
    'AgeCategory':[42, 57, 67, 77, 85],
    'BMI':[16.60, 20.34, 26.58, 24.21, 23.71],
    'KidneyDisease':['no', 'yes', 'no', 'yes', 'no']
}
df = pd.DataFrame(data)

#convert KidneyDisease into numeric values
df['KidneyDisease'] = df['KidneyDisease'].map({'no': 0, 'yes': 1})

#calculate correlation matrix
corr = df.corr()

#plot the heatmap
sb.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')

#add title
plt.title('Correlation Matrix for AgeCategory, BMI and KidneyDisease')

#display
plt.show()
```

### this is a statisticval view of the subject

```python
#sample data with KidneyDisease as string
data = {
    'AgeCategory':[42, 57, 67, 77, 85],
    'BMI':[16.60, 20.34, 26.58, 24.21, 23.71],
    'KidneyDisease':['no', 'yes', 'no', 'yes', 'no']
}
df = pd.DataFrame(data)

#convert KidneyDisease into numeric values
df['KidneyDisease'] = df['KidneyDisease'].map({'no': 0, 'yes': 1})

#calculate correlation matrix
corr_matrix = df.corr()

#print the outcome
print(corr_matrix)
```

```python
#distribution of a continuous variable like BMI
sb.histplot(df['BMI'], kde=True)
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
#probability of a smoker having skin cancer
Smokers = df[df['Smoking'] == 'Yes']
probability_skin_cancer = Smokers['SkinCancer'].value_counts(normalize=True)
print(f"Probability of a smoker having skin cancer:{probability_skin_cancer[1]*100:.2f}%")
```

- Here is a scatter diagram representing the relationship between two continuous variables

```python
#relationship between two continuous variables ie BMI and Age
sb.scatterplot(x='BMI', y='AgeCategory', data=df)
plt.title('The Relationship between BMI and AgeCategory')
plt.xlabel('BMI')
plt.ylabel('AgeCategory')
plt.show()
```

```python
#advice based on HeartDisease  and  GenHealth
gen_health_counts = df[df['HeartDisease']=='Yes']['GenHealth'].value_counts()
gen_health_counts.plot(kind='bar')
plt.title('GenHealth status of patients with HeartDisease')
plt.xlabel('GenHealth')
plt.ylabel('Number of patients')
plt.show()
```
